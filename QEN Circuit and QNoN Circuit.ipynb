## QEN Circuit

```python
n_qubits = 8
qen_circ = QuantumCircuit(n_qubits)
angles1 = [1] * n_qubits
angles2 = [2] * n_qubits
print(angles)

for i in range(n_qubits):
    qen_circ.ry(theta=angles1[i], qubit=i)
    qen_circ.ry(theta=angles2[i], qubit=i)

qen_circ.cnot(0, 1)
qen_circ.cnot(3, 2)
qen_circ.cnot(4, 5)
qen_circ.cnot(7, 6)

qen_circ.ry(theta=3, qubit = 1)
qen_circ.ry(theta=3, qubit = 2)
qen_circ.ry(theta=3, qubit = 5)
qen_circ.ry(theta=3, qubit = 6)

qen_circ.cnot(1, 2)
qen_circ.cnot(6, 5)

qen_circ.ry(theta=4, qubit = 2)
qen_circ.ry(theta=4, qubit = 5)

qen_circ.cnot(2, 5)

qen_circ.ry(5, qubit=5)


qen_circ.draw(output='mpl')
```

## QNoN Circuit

```python
n_qubits = 12
qnon_circ = QuantumCircuit(n_qubits)
angles1 = [1] * n_qubits
angles2 = [2] * n_qubits
print(angles)

for i in range(n_qubits):
    qnon_circ.ry(theta=angles1[i], qubit=i)
    qnon_circ.ry(theta=angles2[i], qubit=i)

qnon_circ.cnot(0, 1)
qnon_circ.cnot(3, 2)
qnon_circ.cnot(4, 5)
qnon_circ.cnot(7, 6)
qnon_circ.cnot(8, 9)
qnon_circ.cnot(11, 10)

qnon_circ.ry(theta=3, qubit = 1)
qnon_circ.ry(theta=3, qubit = 2)
qnon_circ.ry(theta=3, qubit = 5)
qnon_circ.ry(theta=3, qubit = 6)
qnon_circ.ry(theta=3, qubit = 9)
qnon_circ.ry(theta=3, qubit = 10)

qnon_circ.cnot(1, 2)
qnon_circ.cnot(6, 5)
qnon_circ.cnot(10, 9)

qnon_circ.ry(theta=4, qubit = 2)
qnon_circ.ry(theta=4, qubit = 5)
qnon_circ.ry(theta=4, qubit = 9)


qnon_circ.cnot(2, 5)
qnon_circ.ry(5, qubit=5)
qnon_circ.cnot(5, 9)
qnon_circ.ry(6, qubit=9)


qnon_circ.draw(output='mpl')
```
